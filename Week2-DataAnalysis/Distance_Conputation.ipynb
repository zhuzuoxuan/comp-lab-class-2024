```python

trajectory_file = './Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc'
topology_file = './Data/1hz3.pdb'

# Import necessary libraries
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

# Part a: Read the MD trajectory of beta amyloid and the gro file as the structure
trajectory_file = './Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc'  
structure_file = './Data/1hz3_T310.start.nowater.gro'    

# Load the trajectory using mdtraj, including the structure file
traj = md.load(trajectory_file, top=structure_file)

# Part b: Compute the end-to-end distance and radius of gyration at each time
# Assuming end-to-end distance is between the first and last residue

# Compute end-to-end distance (between the first and last atom in the topology)
end_to_end_distances = md.compute_distances(traj, [[0, traj.n_atoms - 1]])[:, 0]

# Compute radius of gyration
radius_of_gyration = md.compute_rg(traj)

# Part c: Plot end-to-end distance and radius of gyration vs time on the same plot
time = traj.time  # Time in picoseconds

plt.figure(figsize=(10, 6))
plt.plot(time, end_to_end_distances, label='End-to-End Distance (nm)', color='b')
plt.plot(time, radius_of_gyration, label='Radius of Gyration (nm)', color='r')
plt.xlabel('Time (ps)')
plt.ylabel('Distance (nm)')
plt.title('End-to-End Distance and Radius of Gyration vs Time')
plt.legend()
plt.grid(True)
plt.show()

# Part d: Plot normalized histograms of end-to-end distance and radius of gyration
plt.figure(figsize=(12, 5))

# Histogram for end-to-end distance
plt.subplot(1, 2, 1)
plt.hist(end_to_end_distances, bins=30, density=True, color='b', alpha=0.7)
plt.xlabel('End-to-End Distance (nm)')
plt.ylabel('Probability Density')
plt.title('Normalized Histogram of End-to-End Distance')

# Histogram for radius of gyration
plt.subplot(1, 2, 2)
plt.hist(radius_of_gyration, bins=30, density=True, color='r', alpha=0.7)
plt.xlabel('Radius of Gyration (nm)')
plt.ylabel('Probability Density')
plt.title('Normalized Histogram of Radius of Gyration')

plt.tight_layout()
plt.show()

```
