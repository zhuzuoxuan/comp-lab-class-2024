```python
import numpy as np
import matplotlib.pyplot as plt

# Function to load data from the xvg file
def load_data(file_path):
    # Load data into a NumPy array, skipping the first 40 lines
    data = np.loadtxt(file_path, skiprows=40)
    return data

# File path
file_path = 'Data/1hz3_T310.run.25000000.energy.xvg'

# Load data
data = load_data(file_path)

# Extract columns
time = data[:, 0] * 1e-9  # Convert time to nanoseconds
potential = data[:, 1]
kinetic_energy = data[:, 2]
total_energy = data[:, 3]
temperature = data[:, 4]
pressure = data[:, 5]
volume = data[:, 6]

# Plot each quantity separately
fig, axs = plt.subplots(3, 2, figsize=(14, 12))

# Define titles and labels
titles = ['Potential', 'Kinetic Energy', 'Total Energy', 'Temperature', 'Pressure', 'Volume']
data_columns = [potential, kinetic_energy, total_energy, temperature, pressure, volume]

for i, ax in enumerate(axs.flat):
    ax.plot(time, data_columns[i])
    ax.set_title(titles[i])
    ax.set_xlabel('Time (ns)')
    ax.set_ylabel(titles[i])
    ax.grid(True)

plt.tight_layout()
plt.show()

# Combined plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:red'
ax1.set_xlabel('Time (ns)')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(time, temperature, color=color, label='Temperature')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # Create a second y-axis that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('Kinetic Energy', color=color)
ax2.plot(time, kinetic_energy, color=color, label='Kinetic Energy')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # Ensure everything fits without overlap
plt.title('Kinetic Energy and Temperature')
plt.show()
```
