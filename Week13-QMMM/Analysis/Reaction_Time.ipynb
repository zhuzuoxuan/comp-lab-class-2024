```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path
colvar_file = '../Outputs/monitor/MONITOR-COLVAR.metadynLog'

# Load the data, assuming it's whitespace-delimited
colvar_data = pd.read_csv(colvar_file, delim_whitespace=True, header=None)

# Assign column names for clarity
colvar_data.columns = ['Time_fs', 'Reaction_Coordinate', 'Col3', 'Col4', 'Col5', 'Col6']

# Plot the reaction coordinate vs time
plt.figure(figsize=(10, 6))
plt.plot(colvar_data['Time_fs'], colvar_data['Reaction_Coordinate'], label="Reaction Coordinate", linestyle='-', marker='.')

# Add labels, title, and legend
plt.xlabel("Time (fs)")
plt.ylabel("Reaction Coordinate")
plt.title("Reaction Coordinate vs Time")
plt.legend()
plt.grid()

# Save the figure
plt.savefig('Reaction_Coordinate_vs_Time.png', dpi=300)
plt.show()
```

```python
# Define the file path for metad1_initmonitor
init_monitor_file = '../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog'

# Load the data, assuming it's whitespace-delimited
init_monitor_data = pd.read_csv(init_monitor_file, delim_whitespace=True, header=None)

# Assign column names for clarity (if known structure is similar to MONITOR-COLVAR.metadynLog)
init_monitor_data.columns = ['Time_fs', 'Reaction_Coordinate', 'Col3', 'Col4', 'Col5', 'Col6']

# Display the range of Reaction Coordinate values
reaction_coord_values = init_monitor_data['Reaction_Coordinate']
print(f"Reaction Coordinate values from {init_monitor_file}:")
print(reaction_coord_values.describe())

# Check for transitions
if (reaction_coord_values.min() < 0) and (reaction_coord_values.max() > 0):
    print("A reaction is observed: Transition from negative to positive or positive to negative values.")
else:
    print("No reaction observed: No transition between positive and negative values.")

# Plot the reaction coordinate for this file
plt.figure(figsize=(10, 6))
plt.plot(init_monitor_data['Time_fs'], init_monitor_data['Reaction_Coordinate'], label="Reaction Coordinate (Init Monitor)", linestyle='-', marker='.')

# Add labels, title, and legend
plt.xlabel("Time (fs)")
plt.ylabel("Reaction Coordinate")
plt.title("Reaction Coordinate vs Time (Init Monitor)")
plt.legend()
plt.grid()

# Save the figure
plt.savefig('Reaction_Coordinate_vs_Time_Init_Monitor.png', dpi=300)
plt.show()
```
