```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Define column names based on file structure
columns = ["Step", "Time_fs", "Kin_a.u.", "Temp_K", "Pot_a.u.", "Cons_Qty_a.u.", "UsedTime_s"]

# Load NPT and NVT data
npt_file = '../Outputs/NPT-1.ener'
nvt_file = '../Outputs/NVT-1.ener'

npt_data = pd.read_csv(npt_file, delim_whitespace=True, comment='#', names=columns)
nvt_data = pd.read_csv(nvt_file, delim_whitespace=True, comment='#', names=columns)

# Plot Potential Energy vs Time for NPT and NVT
plt.figure(figsize=(10, 6))
plt.plot(npt_data["Time_fs"], npt_data["Pot_a.u."], label="NPT Potential Energy", linestyle='-', marker='.')
plt.plot(nvt_data["Time_fs"], nvt_data["Pot_a.u."], label="NVT Potential Energy", linestyle='-', marker='.')

# Labels and title
plt.xlabel("Time (fs)")
plt.ylabel("Potential Energy (a.u.)")
plt.title("Potential Energy vs Time (NPT and NVT)")
plt.legend()
plt.grid()

# Save the figure
plt.savefig('Potential_Energy_vs_Time_NPT_NVT.png', dpi=300)
plt.show()
```

```python
# Load MONITOR-1.ener file
monitor_file = '../Outputs/monitor/MONITOR-1.ener'
monitor_data = pd.read_csv(monitor_file, delim_whitespace=True, comment='#', names=columns)

# Plot Potential Energy vs Time for MONITOR-1.ener
plt.figure(figsize=(10, 6))
plt.plot(monitor_data["Time_fs"], monitor_data["Pot_a.u."], label="QM/MM Potential Energy", linestyle='-', marker='.')

# Labels and title
plt.xlabel("Time (fs)")
plt.ylabel("Potential Energy (a.u.)")
plt.title("Potential Energy vs Time (QM/MM)")
plt.legend()
plt.grid()

# Save the figure
plt.savefig('Potential_Energy_vs_Time_QMMM.png', dpi=300)
plt.show()
```
