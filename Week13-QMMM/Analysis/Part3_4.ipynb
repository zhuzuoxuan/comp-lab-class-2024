```python
import pandas as pd
import matplotlib.pyplot as plt

# ==================== 1.  ====================
# 
colvar_file = '../Outputs/monitor/MONITOR-COLVAR.metadynLog'  # COLVAR
metadynamics_inp_file = '../Outputs/monitor/monitor.inp'                # metadynamics

#  COLVAR  ()
colvar_data = pd.read_csv(colvar_file, delim_whitespace=True, header=None)
colvar_data.columns = ['Time_fs', 'Reaction_Coordinate', 'Col3', 'Col4', 'Col5', 'Col6']

#  (fs)  (ps)
colvar_data['Time_ps'] = colvar_data['Time_fs'] / 1000.0

# ==================== 2.  metadynamics  ====================
hill_height, hill_width, hill_frequency = None, None, None  # 

#  metadynamics.inp 
with open(metadynamics_inp_file, 'r') as file:
    for line in file:
        if 'WW' in line:  # Hill Height
            hill_height = line.split()[1].strip()
        elif 'WIDTH' in line:  # Hill Width
            hill_width = line.split()[1].strip()
        elif 'NT_HILLS' in line:  # Hill Frequency
            hill_frequency = line.split()[1].strip()

# ==================== 3.  ====================
reaction_coord_formula = r"Reaction Coordinate = -1 * [(d(5663, 5675)) - (d(5670, 5671))]"

# ==================== 4.  ====================
plt.figure(figsize=(10, 6))
plt.plot(colvar_data['Time_ps'], colvar_data['Reaction_Coordinate'],
         label=(f"Reaction Coordinate (QM/MM)\n{reaction_coord_formula}\n"
                f"Hill Height: {hill_height} kcal/mol, "
                f"Hill Width: {hill_width}, Hill Frequency: {hill_frequency} fs"),
         linestyle='-', marker='.')

# 
plt.xlabel("Time (ps)")
plt.ylabel("Reaction Coordinate")
plt.title("Reaction Coordinate vs Time (QM/MM)")
plt.legend()
plt.grid()

# 
plt.savefig('Reaction_Coordinate_vs_Time_QMMM.png', dpi=300)
plt.show()
```

```python
# ==================== 5.  ====================
init_monitor_file = '../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog'  # 

# 
init_monitor_data = pd.read_csv(init_monitor_file, delim_whitespace=True, header=None)
init_monitor_data.columns = ['Time_fs', 'Reaction_Coordinate', 'Col3', 'Col4', 'Col5', 'Col6']

#  ps
init_monitor_data['Time_ps'] = init_monitor_data['Time_fs'] / 1000.0

# 
reaction_occurred = ((init_monitor_data['Reaction_Coordinate'] < 0).any() and
                     (init_monitor_data['Reaction_Coordinate'] > 0).any())

if reaction_occurred:
    print("A reaction is observed: Transition from positive to negative or negative to positive values of the CV.")
else:
    print("No reaction observed: The CV does not transition between positive and negative values.")

# ==================== 6.  ====================
plt.figure(figsize=(10, 6))
plt.plot(init_monitor_data['Time_ps'], init_monitor_data['Reaction_Coordinate'],
         label="Initial Reaction Coordinate (metad1_initmonitor)", linestyle='-', marker='.')

# 
plt.xlabel("Time (ps)")
plt.ylabel("Reaction Coordinate")
plt.title("Initial Reaction Coordinate vs Time")
plt.legend()
plt.grid()

# 
plt.savefig('Initial_Reaction_Coordinate_vs_Time.png', dpi=300)
plt.show()
```
