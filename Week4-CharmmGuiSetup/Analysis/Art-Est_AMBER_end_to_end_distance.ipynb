```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

# 
data = []

#  .xvg 
with open('end_to_end_dist.xvg', 'r') as file:
    for line in file:
        #  @  # 
        if line.startswith(('@', '#')):
            continue
        # 
        data.append([float(x) for x in line.split()])

#  pandas DataFrame
end_to_end_data = pd.DataFrame(data, columns=['Time', 'Distance'])

# 
plt.figure(figsize=(8, 6))
plt.plot(end_to_end_data['Time'], end_to_end_data['Distance'], color='b', label='End-to-End Distance')
plt.xlabel('Time (ps)')
plt.ylabel('Distance (nm)')
plt.title('End-to-End Distance vs Time')
plt.legend()
plt.grid(True)
plt.savefig('end_to_end_distance_vs_time.png') 
plt.show()
```
