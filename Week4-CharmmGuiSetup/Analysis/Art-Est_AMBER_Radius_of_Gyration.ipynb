```python
import matplotlib.pyplot as plt
import numpy as np

# Load data from the .xvg file (assuming two columns: time and radius of gyration)
data = np.loadtxt('gyrate.xvg', comments=['@', '#'])

# Extract time and radius of gyration
time = data[:, 0]
radius_of_gyration = data[:, 1]

# Plot
plt.figure()
plt.plot(time, radius_of_gyration, label='Radius of Gyration', color='orange')
plt.xlabel('Time (ps)')
plt.ylabel('Radius of Gyration ()')
plt.title('Radius of Gyration vs Time')
plt.legend()
plt.savefig('Radius_of_Gyration.png') 
plt.show()
```
