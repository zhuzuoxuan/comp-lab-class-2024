```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Function to read potential energy from an XVG file
def read_xvg(file_path):
    data = []
    with open(file_path, 'r') as f:
        for line in f:
            if line.startswith('@') or line.startswith('#'):
                continue  # Skip metadata lines
            data.append([float(x) for x in line.split()])
    return np.array(data)

# Read potential energy data for each temperature
energy_T300 = read_xvg('T300/potential_energy.xvg')[:, 1]  # Column 1 is time, Column 2 is energy
energy_T400 = read_xvg('T400/potential_energy.xvg')[:, 1]
energy_T600 = read_xvg('T600/potential_energy.xvg')[:, 1]

# Plot histograms of the potential energies
plt.figure(figsize=(10, 6))
bins = 50  # Adjust the number of bins as needed

plt.hist(energy_T300, bins=bins, alpha=0.5, label='T = 300 K', color='blue', density=True)
plt.hist(energy_T400, bins=bins, alpha=0.5, label='T = 400 K', color='green', density=True)
plt.hist(energy_T600, bins=bins, alpha=0.5, label='T = 600 K', color='red', density=True)

plt.xlabel('Potential Energy (kJ/mol)')
plt.ylabel('Probability Density')
plt.title('Histogram of Potential Energy for Different Temperatures')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
