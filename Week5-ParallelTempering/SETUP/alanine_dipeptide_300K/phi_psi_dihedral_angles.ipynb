```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np

# Load the trajectory (e.g., adp_T300.xtc) and topology (e.g., adp.gro)
trajectory = md.load('adp_T300.xtc', top='adp.gro')

# Compute phi and psi angles
phi_indices, phi_angles = md.compute_phi(trajectory)
psi_indices, psi_angles = md.compute_psi(trajectory)

# Convert angles from radians to degrees
phi_angles_degrees = np.degrees(phi_angles)
psi_angles_degrees = np.degrees(psi_angles)

# Extract time data in picoseconds (ps)
time = trajectory.time

# Plot Phi angles vs. time
plt.figure(figsize=(10, 5))
plt.plot(time, phi_angles_degrees[:, 0], label='Phi Angle', color='blue')
plt.xlabel('Time (ps)')
plt.ylabel('Phi Angle (degrees)')
plt.title('Phi Angle vs Time')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()

output_filename = "Phi_Angle_Time"
plt.savefig(output_filename, dpi=300)
plt.close()

# Plot Psi angles vs. time
plt.figure(figsize=(10, 5))
plt.plot(time, psi_angles_degrees[:, 0], label='Psi Angle', color='red')
plt.xlabel('Time (ps)')
plt.ylabel('Psi Angle (degrees)')
plt.title('Psi Angle vs Time')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()

output_filename = "Psi_Angle_Time"
plt.savefig(output_filename, dpi=300)
plt.close()

```
