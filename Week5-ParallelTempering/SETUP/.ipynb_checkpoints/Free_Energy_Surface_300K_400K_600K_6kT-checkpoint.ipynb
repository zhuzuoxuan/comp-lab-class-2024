```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt
import os

# Define a function to compute and plot the free energy surface
def compute_and_save_fes(traj_path, top_path, temperature, output_filename, bins=50):
    """
    Compute the free energy surface (FES) in phi-psi space for a given temperature.
    
    Parameters:
    traj_path : str
        Path to the trajectory file (.xtc).
    top_path : str
        Path to the topology file (.gro).
    temperature : float
        Temperature in Kelvin.
    output_filename : str
        Filename to save the plot.
    bins : int
        Number of bins for the histogram.
    """
    # Load the trajectory
    trajectory = md.load(traj_path, top=top_path)

    # Compute phi and psi angles
    phi_indices, phi_angles = md.compute_phi(trajectory)
    psi_indices, psi_angles = md.compute_psi(trajectory)

    # Convert angles to degrees for easier interpretation
    phi_angles_deg = np.degrees(phi_angles[:, 0])
    psi_angles_deg = np.degrees(psi_angles[:, 0])

    # Create a 2D histogram of phi and psi angles
    hist, x_edges, y_edges = np.histogram2d(
        phi_angles_deg, psi_angles_deg, bins=bins, range=[[-180, 180], [-180, 180]]
    )

    # Calculate free energy surface: F(phi, psi) = -k_B * T * ln(histogram)
    k_B = 0.008314  # Boltzmann constant in kJ/(mol*K)

    # Avoid taking the log of zero by adding a small value to the histogram
    hist += 1e-10
    fes = -k_B * temperature * np.log(hist)

    # Normalize the free energy to set the minimum to zero
    fes -= np.min(fes)

    # Set the maximum free energy to 6 kT for visualization
    max_free_energy = 6 * k_B * temperature  # Calculate 6 kT in kJ/mol
    fes[fes > max_free_energy] = max_free_energy

    # Calculate the bin centers for plotting
    x_bin_centers = 0.5 * (x_edges[:-1] + x_edges[1:])
    y_bin_centers = 0.5 * (y_edges[:-1] + y_edges[1:])

    # Plot the capped free energy surface and save as an image
    plt.figure(figsize=(8, 6))
    extent = [x_bin_centers[0], x_bin_centers[-1], y_bin_centers[0], y_bin_centers[-1]]
    im = plt.imshow(fes.T, extent=extent, origin='lower', aspect='auto', cmap='viridis')
    plt.colorbar(im, label=f'F(φ, ψ) - min(F) (kJ/mol), capped at 6 kT')
    plt.xlabel('Phi (degrees)')
    plt.ylabel('Psi (degrees)')
    plt.title(f'Free Energy Surface (FES) at T = {temperature} K, max 6 kT')
    plt.grid(True)
    plt.tight_layout()

    # Show the plot before saving
    plt.show()


    # Save the plot
    plt.savefig(output_filename, dpi=300)
    plt.close()
    print(f'FES for T = {temperature} K saved as {output_filename}')

# Define paths and parameters for each temperature
temperatures = [300, 400, 600]
traj_paths = ['T300/adp_exchange3temps.xtc', 'T400/adp_exchange3temps.xtc', 'T600/adp_exchange3temps.xtc']
top_path = 'T300/adp.gro'  # Adjust this if the topology files differ for each temperature
output_files = ['FES_T300.png', 'FES_T400.png', 'FES_T600.png']

# Generate and save the FES for each temperature
for temp, traj_path, output_file in zip(temperatures, traj_paths, output_files):
    compute_and_save_fes(traj_path, top_path, temp, output_file)
```
