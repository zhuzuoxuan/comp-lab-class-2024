```python
import numpy as np
import matplotlib.pyplot as plt

# 读取数据函数
def read_xvg(filename):
    data = []
    with open(filename, 'r') as file:
        for line in file:
            # 跳过以@或#开头的行
            if not (line.startswith('@') or line.startswith('#')):
                # 分割行并将其转换为浮点数
                parts = line.split()
                if len(parts) >= 2:  # 确保有足够的列
                    data.append([float(parts[0]), float(parts[1])])
    return np.array(data)


# 读取数据
filename = './temperature.xvg'  # 请替换为你的文件名
data = read_xvg(filename)

# 提取x和y值
x = data[:, 0]  # 第一列：Energy Minimization Step
y = data[:, 1]  # 第二列：Potential Energy (kJ/mol)

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Temperature', color='blue')
plt.title('Temperature')
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.grid()
plt.legend()
plt.savefig('Trp-cage_NVT_Tempeature.png')
plt.show()
```
