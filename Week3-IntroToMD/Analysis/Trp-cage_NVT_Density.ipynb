```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data file
file_path = "./Trp-cage_NVT_Density.xvg"

# Read the file, skipping initial lines starting with "#" or "@" which are comments
with open(file_path, 'r') as file:
    lines = file.readlines()

# Filter out comments and extract data lines
data_lines = [line for line in lines if not line.startswith(('#', '@'))]

# Convert the data lines into a pandas DataFrame
data = pd.DataFrame([list(map(float, line.split())) for line in data_lines], columns=["Time (ps)", "Density (g/cm^3)"])

# Calculate the running average of the density data
window_size = 10  # Define the window size for the running average
data['Running Average'] = data['Density (g/cm^3)'].rolling(window=window_size).mean()

# Plotting the data with running average
plt.figure(figsize=(10, 6))
plt.plot(data["Time (ps)"], data["Density (g/cm^3)"], label='Density over Time', color='b')
plt.plot(data["Time (ps)"], data["Running Average"], label=f'Running Average ({window_size} points)', color='r')
plt.title('Density')
plt.xlabel('Time (ps)')
plt.ylabel('Density (g/cm^3)')
plt.legend()
plt.grid(True)

# Save and display the plot
plt.savefig('Trp-cage_NVT_Density.png')
plt.show()

```
