```python
import numpy as np
import matplotlib.pyplot as plt

# Read in the COLVAR file using numpy
data = np.loadtxt('COLVAR')

# Assuming the COLVAR file has three columns: time, phi, and psi
# If your COLVAR file has different formatting, you may need to adjust this.
time = data[:, 0]  # First column: time
phi = data[:, 1]   # Second column: phi angle
psi = data[:, 2]   # Third column: psi angle

# Plot the phi angle over time
plt.figure(figsize=(8, 5))
plt.plot(time, phi, label='Phi', color='blue')
plt.xlabel('Time (ps)')
plt.ylabel('Dihedral Angle (degrees)')
plt.title('Phi Angle Over Time')
plt.grid()
plt.legend()
plt.tight_layout()
plt.savefig('alanine_dipeptide_B_Phi_Angle_Over_Time.png')
plt.show()

# Plot the psi angle over time
plt.figure(figsize=(8, 5))
plt.plot(time, psi, label='Psi', color='red')
plt.xlabel('Time (ps)')
plt.ylabel('Dihedral Angle (degrees)')
plt.title('Psi Angle Over Time')
plt.grid()
plt.legend()
plt.tight_layout()
plt.savefig('alanine_dipeptide_B_Psi_Angle_Over_Time.png')
plt.show()

# Optional: If you want to plot phi vs psi in a Ramachandran-style plot:
plt.figure(figsize=(6, 6))
plt.scatter(phi, psi, c=time, cmap='viridis', alpha=0.7)
plt.colorbar(label='Time (ps)')
plt.xlabel('Phi (degrees)')
plt.ylabel('Psi (degrees)')
plt.title('Ramachandran Plot')
plt.grid()
plt.tight_layout()
plt.savefig('alanine_dipeptide_B_Ramachandran_Plot.png')
plt.show()
```
