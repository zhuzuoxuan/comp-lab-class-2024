```python
import matplotlib.pyplot as plt
import pandas as pd

# 读取 COLVAR 文件，跳过前两行注释
file_path = "COLVAR"
try:
    # 读取数据，从第三行开始，指定列名
    data = pd.read_csv(
        file_path,
        delim_whitespace=True,
        comment='#',
        names=["time", "phi", "psi", "metad_bias"]
    )
except Exception as e:
    print(f"Error reading the COLVAR file: {e}")
    exit()

# 检查是否成功读取数据
if data.empty:
    print("No data was loaded. Please check the COLVAR file format.")
    exit()

# 绘制 phi 随时间变化的图像
plt.figure(figsize=(8, 5))
plt.plot(data["time"], data["phi"], label=r'$\phi$ angle', linewidth=1)
plt.xlabel("Time (ps)")
plt.ylabel(r'$\phi$ angle (radians)')
plt.title(r'Time evolution of $\phi$ angle')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('metadynamics_calculation_Time_evolution_of_phi_angle')
plt.show()

# 绘制 psi 随时间变化的图像
plt.figure(figsize=(8, 5))
plt.plot(data["time"], data["psi"], label=r'$\psi$ angle', color='orange', linewidth=1)
plt.xlabel("Time (ps)")
plt.ylabel(r'$\psi$ angle (radians)')
plt.title(r'Time evolution of $\psi$ angle')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('metadynamics_calculation_Time_evolution_of_psi_angle')
plt.show()

# 绘制 Ramachandran 图，使用时间作为颜色渐变
plt.figure(figsize=(8, 8))
sc = plt.scatter(data["phi"], data["psi"], c=data["time"], cmap='viridis', s=1, alpha=0.7)
plt.xlabel(r'$\phi$ angle (radians)')
plt.ylabel(r'$\psi$ angle (radians)')
plt.title(r'Ramachandran Plot with Time')
plt.colorbar(sc, label='Time (ps)')  # 添加颜色条，表示时间
plt.grid(True)
plt.tight_layout()
plt.xlim(-3.14, 3.14)  # 限制 x 轴的范围为 -pi 到 pi
plt.ylim(-3.14, 3.14)  # 限制 y 轴的范围为 -pi 到 pi
plt.savefig('metadynamics_calculation_Ramachandran_Plot')
plt.show()
```
