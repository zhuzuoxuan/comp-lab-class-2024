```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#  custom_fes.dat  phi  file.free 
file_path = "custom_fes.dat"  # 
try:
    # 
    data = pd.read_csv(
        file_path,
        delim_whitespace=True,
        comment='#',
        names=["phi", "free_energy", "der_phi"]
    )
except Exception as e:
    print(f"Error reading the FES file: {e}")
    exit()

# 
if data.empty:
    print("No data was loaded. Please check the FES file format.")
    exit()

#  (FES)
plt.figure(figsize=(8, 6))
plt.plot(data["phi"], data["free_energy"], label='FES', color='blue', linewidth=2)
plt.xlabel(r'$\phi$ angle (radians)')
plt.ylabel('Free Energy (kJ/mol)')
plt.title('Free Energy Surface (FES)')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.xlim(-np.pi, np.pi)  #  x  -pi  pi
plt.savefig('Estimate_of_FE_as_function_of_dihedral_phi.png')
plt.show()
```
