```python
import matplotlib.pyplot as plt
import pandas as pd

#  HILLS 
file_path = "HILLS"
try:
    # 
    data = pd.read_csv(
        file_path,
        delim_whitespace=True,
        comment='#',
        names=["time", "phi", "sigma_phi", "height", "biasf"]
    )
except Exception as e:
    print(f"Error reading the HILLS file: {e}")
    exit()

# 
if data.empty:
    print("No data was loaded. Please check the HILLS file format.")
    exit()

# 
plt.figure(figsize=(8, 5))
plt.plot(data["time"], data["height"], label='Gaussian Height', linewidth=1, color='blue')
plt.xlabel("Time (ps)")
plt.ylabel("Gaussian Height (kJ/mol)")
plt.title("Decrease of Gaussian Height during Well-Tempered Metadynamics")
plt.grid(True)
plt.tight_layout()
plt.savefig('metadynamics_calculation_Decrease_of_Gaussian_Height')
plt.show()

#  phi 
plt.figure(figsize=(8, 5))
plt.plot(data["time"], data["phi"], label=r'$\phi$ position', linewidth=1, color='green')
plt.xlabel("Time (ps)")
plt.ylabel(r'$\phi$ (radians)')
plt.title("Position of $\phi$ during Well-Tempered Metadynamics")
plt.grid(True)
plt.tight_layout()
plt.savefig('metadynamics_calculation_Position_of_phi_during_Well-Tempered_Metadynamics')
plt.show()
```
