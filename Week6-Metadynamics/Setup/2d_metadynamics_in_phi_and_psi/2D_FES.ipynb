```python
import numpy as np
import matplotlib.pyplot as plt

# 指定要处理的 FES 文件及其对应步数
fes_files = [
    ("fes_%i.dat0.dat", 0),
    ("fes_%i.dat10.dat", 100000),
    ("fes_%i.dat20.dat", 200000),
    ("fes_%i.dat30.dat", 300000),
    ("fes_%i.dat40.dat", 400000),
    ("fes_%i.dat50.dat", 500000),
    ("fes_%i.dat60.dat", 600000)
]

# 遍历指定的文件并绘制
for fes_file, step in fes_files:
    # 读取数据 (phi, psi, Free Energy)
    data = np.loadtxt(fes_file)
    phi = data[:, 0]
    psi = data[:, 1]
    free_energy = data[:, 2]

    # 绘制 2D 等高图
    plt.figure(figsize=(8, 6))
    plt.tricontourf(phi, psi, free_energy, levels=20, cmap='viridis')
    plt.colorbar(label="Free Energy (kJ/mol)")
    plt.xlabel("Phi (radians)")
    plt.ylabel("Psi (radians)")
    plt.title(f"FES at step {step}")

    # 保存图片
    plt.savefig(f"FES_step_{step}.png")

    # 显示图片
    plt.show()
```
