```python
import numpy as np
import matplotlib.pyplot as plt

#  (sum_hills  fes.dat )
data = np.loadtxt("fes.dat")  # 

# phi  psi 
phi = np.unique(data[:, 0])
psi = np.unique(data[:, 1])
free_energy = data[:, 2].reshape(len(psi), len(phi))  # 2D

# 
plt.figure(figsize=(10, 8))
im = plt.imshow(
    free_energy, 
    extent=[phi.min(), phi.max(), psi.min(), psi.max()],
    origin='lower', 
    cmap='viridis', 
    aspect='auto'
)

# 
plt.colorbar(im, label="Free Energy (kJ/mol)")
plt.xlabel("Phi (radians)")
plt.ylabel("Psi (radians)")
plt.title("Free Energy Surface")

# 
contour = plt.contour(
    phi, psi, free_energy, levels=15, colors='white', linewidths=0.8
)
plt.clabel(contour, inline=True, fontsize=8, fmt="%.1f")

# 
plt.savefig("Free_Energy_Surface.png", dpi=300)
plt.show()
```
