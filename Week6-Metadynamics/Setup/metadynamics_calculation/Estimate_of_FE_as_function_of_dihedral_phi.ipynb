```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 读取 custom_fes.dat 文件，假设文件包含 phi 和 file.free 的列
file_path = "custom_fes.dat"  # 根据实际输出文件名修改
try:
    # 读取数据，跳过注释行并指定列名
    data = pd.read_csv(
        file_path,
        delim_whitespace=True,
        comment='#',
        names=["phi", "free_energy", "der_phi"]
    )
except Exception as e:
    print(f"Error reading the FES file: {e}")
    exit()

# 检查是否成功读取数据
if data.empty:
    print("No data was loaded. Please check the FES file format.")
    exit()

# 绘制自由能面 (FES)
plt.figure(figsize=(8, 6))
plt.plot(data["phi"], data["free_energy"], label='FES', color='blue', linewidth=2)
plt.xlabel(r'$\phi$ angle (radians)')
plt.ylabel('Free Energy (kJ/mol)')
plt.title('Free Energy Surface (FES)')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.xlim(-np.pi, np.pi)  # 设置 x 轴范围为 -pi 到 pi
plt.savefig('Estimate_of_FE_as_function_of_dihedral_phi.png')
plt.show()
```
