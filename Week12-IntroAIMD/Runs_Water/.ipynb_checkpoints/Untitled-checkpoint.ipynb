```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path to the energy file
energy_file = "water_aimd_1ns_every0.5fs-1.ener"

# Load the energy data from the file
try:
    # Read the data while skipping the comment line at the top
    data = pd.read_csv(
        energy_file,
        delim_whitespace=True,
        comment='#',
        names=[
            "Step",
            "Time (fs)",
            "Kin Energy (a.u.)",
            "Temperature (K)",
            "Pot Energy (a.u.)",
            "Cons Qty (a.u.)",
            "Used Time (s)"
        ]
    )
except FileNotFoundError:
    print(f"File {energy_file} not found. Ensure the file is in the working directory.")
    raise

# Plot Potential Energy vs. Time
plt.figure(figsize=(10, 6))
plt.plot(data["Time (fs)"], data["Pot Energy (a.u.)"], label="Potential Energy", color="blue")
plt.title("Potential Energy vs Time")
plt.xlabel("Time (fs)")
plt.ylabel("Potential Energy (a.u.)")
plt.grid(True)
plt.legend()
plt.savefig("potential_energy_vs_time.png")  # Save plot to a file
plt.show()

# Plot Kinetic Energy vs. Time
plt.figure(figsize=(10, 6))
plt.plot(data["Time (fs)"], data["Kin Energy (a.u.)"], label="Kinetic Energy", color="green")
plt.title("Kinetic Energy vs Time")
plt.xlabel("Time (fs)")
plt.ylabel("Kinetic Energy (a.u.)")
plt.grid(True)
plt.legend()
plt.savefig("kinetic_energy_vs_time.png")  # Save plot to a file
plt.show()

# Plot Total Energy (Potential + Kinetic) vs. Time
total_energy = data["Pot Energy (a.u.)"] + data["Kin Energy (a.u.)"]
plt.figure(figsize=(10, 6))
plt.plot(data["Time (fs)"], total_energy, label="Total Energy", color="red")
plt.title("Total Energy vs Time")
plt.xlabel("Time (fs)")
plt.ylabel("Total Energy (a.u.)")
plt.grid(True)
plt.legend()
plt.show()


```
